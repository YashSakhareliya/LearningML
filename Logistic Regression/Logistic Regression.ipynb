```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
dataset = make_classification(n_samples=2000,n_features=8,n_classes=2)
```

```python
type(dataset)
```

```python
dataset
```

```python
dataset
```

```python
df = pd.DataFrame(data=dataset[0])
```

```python
df.head()
```

```python
df.shape
```

```python
df['Output'] = dataset[1]
```

```python
df.shape
```

```python
df.head()
```

```python
# spliting into indepedent and depedent feature
X = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# model training
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
y_pred.shape
```

```python
logistic.predict_proba(X_test)
```

```python
logistic.score(X_test,y_test)
```

```python
# score and confusion metrics and classification report
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
matrix = confusion_matrix(y_test,y_pred)
print(score)
print(matrix)
print(classification_report(y_test,y_pred))
```

## Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    'penalty':['l1', 'l2', 'elasticnet'],
    'C':[100,10,1.0,0.1,0.01],
    'solver':['lbfgs', 'liblinear', 'newton-cg','sag', 'saga']
}
```

```python
type(param)
```

```python
param
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model,param_grid=param,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.get_params
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred_grid)
matrix = confusion_matrix(y_test,y_pred_grid)
print(score)
print(matrix)
print(classification_report(y_test,y_pred_grid))
```

## randmize search cv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv = RandomizedSearchCV(estimator=model,param_distributions=param,scoring='accuracy',cv=5)
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred_randomcv = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred_randomcv)
matrix = confusion_matrix(y_test,y_pred_randomcv)
print(score)
print(matrix)
print(classification_report(y_test,y_pred_randomcv))
```
