```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4]
y = [2, 5, 9, 2]
```

```python
# line chart
plt.subplot(2,2,1)
plt.plot(x)


plt.subplot(2,2,2)
plt.plot(y)


plt.subplot(2,2,3)
plt.plot(x,y)

plt.show()
```

```python
# line chart with all attribute
plt.subplot(2,2,1)
plt.plot(x,linestyle='-.',marker='o', linewidth='3',mew='3')


plt.subplot(2,2,2)
plt.plot(y,ls='--',marker='o',lw=3)


plt.subplot(2,2,3)
plt.plot(x,y)

plt.show()
```

```python
# hsit plot 
plt.hist(x,bins=3)
plt.show()
```

```python
# scatter plot
plt.scatter(x,y)
plt.show() # Basic scatter plot
```

```python
# scatter plot with line
plt.scatter(x,y,ec='b')
plt.plot(x,y,mec='r',mew=4,marker='o')
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```
