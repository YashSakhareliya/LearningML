```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Dataset/Travel_cleaned.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# scaling and column Transfomer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
# train test split
from sklearn.model_selection import train_test_split
X = df.drop(['ProdTaken'],axis=1)
y = df['ProdTaken']
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
num_feature = X.select_dtypes(exclude='object').columns
cat_feature = X.select_dtypes(include='object').columns
```

```python
num_feature
```

```python
cat_feature
```

```python
preprocessor = ColumnTransformer(
    [
        ('ohe',OneHotEncoder(),cat_feature),
        ('scaler',StandardScaler(),num_feature)
    ]
)
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
X_train_transformed
```

```python
X_test_transformed
```

```python
pd.DataFrame(X_train_transformed)
```

```python
# model training
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gradient = GradientBoostingClassifier()
```

```python
gradient.fit(X_train_transformed,y_train)
```

```python
gradient.score(X_test_transformed,y_test)
```

```python
y_pred = gradient.predict(X_test_transformed)
```

```python
# score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
