```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
df = make_blobs(n_samples=1000,n_features=2,centers=3,random_state=10)
```

```python
X = df[0]
y =df[1]
```

```python
type(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
# scaling the feature
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# make cluster
# find the k value 
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1,11):
    kmean =KMeans(n_clusters=k)
    kmean.fit(X_train)
    wcss.append(kmean.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.show()
```

```python
# we select the k = 3
kmean = KMeans(n_clusters=3,init='k-means++')
```

```python
kmean.fit_predict(X_train)
```

```python
# lets visulize the cluster
plt.scatter(X_test[:,0],X_test[:,1],c=kmean.predict(X_test))
```
