```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Travel_cleaned.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

## train test split

```python
X = df.drop(['ProdTaken'],axis=1)
y = df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape,X_test.shape
```

### scaling and Transformation of data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
scaler = StandardScaler()
ohe = OneHotEncoder(drop='first')
```

```python
print(X.select_dtypes(include='object').columns)
X.select_dtypes(exclude='object').columns
```

```python
cat_feature = X.select_dtypes(include='object').columns
num_feature = X.select_dtypes(exclude='object').columns
```

```python
preprocessor = ColumnTransformer(
    [
        ('scaler',scaler,num_feature),
        ('ohe', ohe, cat_feature)
    ]
)
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
X_train_transformed
```

```python
y_train
```

## Model Training

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
Addabost = AdaBoostClassifier()
```

```python
Addabost.fit(X_train_transformed,y_train)
```

```python
Addabost.score(X_test_transformed,y_test)
```

```python
# score and classification report
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred = Addabost.predict(X_test_transformed)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators':[50,60,70,80,90,100],
    'algorithm':['SAMME', 'SAMME.R'],
    'learning_rate':[1.0,0.90,1.10,1.20,1.50,2,0.5]
}
```

```python
param_grid
```

```python
random = RandomizedSearchCV(estimator=AdaBoostClassifier(),scoring='accuracy',verbose=2,cv=5,param_distributions=param_grid,n_jobs=-1)
```

```python
random.fit(X_train_transformed,y_train)
```

```python
random.best_params_
```

```python
y_pred = random.predict(X_test_transformed)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
