## Bag of Words using Porter Stemmer 

```python
import nltk
import re
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
df.head()
```

```python
corpus = []

for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# bag of word
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer(max_features=200)
```

```python
X = vector.fit_transform(corpus).toarray()
```

```python
X.shape
```

## Model Trainig Logistic

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
y = enc.fit_transform(df['lable'])
```

```python
X = pd.DataFrame(X)
y = pd.Series(y,name='Target')
```

```python
X,y
```

```python
y.value_counts()
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape,y_train.shape
```

```python
y_test.value_counts()
```

```python
# model training
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_train_pred = logistic.predict(X_train)
y_test_pred = logistic.predict(X_test)
```

```python
# scoring
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Test data accuracy {}'.format(accuracy_score(y_test,y_test_pred)))
```

```python
print('Train data accuracy {}'.format(accuracy_score(y_train,y_train_pred)))
```

```python
print('Classification report',classification_report(y_test,y_test_pred))
```

## RandForest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random = RandomForestClassifier(verbose=3)
```

```python
random.fit(X_train,y_train)
```

```python
y_train_pred = random.predict(X_train)
y_test_pred = random.predict(X_test)
```

```python
print('RandomForest Train data accuracy {}'.format(accuracy_score(y_train,y_train_pred)))
print('RandomForest Test data accuracy {}'.format(accuracy_score(y_test,y_test_pred)))
```
